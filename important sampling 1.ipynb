```python
import numpy as np
import matplotlib.pyplot as py
import math
```

```python
def func(x):
    return np.exp(-x**2)

def dist(x):
    return (np.exp(1)/(np.exp(1)-1))*np.exp(-x)


func(0)/dist(0) - func(1)/dist(1)
```

```python
a = 0
b = 1
count = 10000
x_data = range(count)
mean_value = []
error_value = []
std = []
x = np.random.uniform(a,b,count)
yx = np.log(np.exp(1)/(np.exp(1)*(1-x)+x))
y = func(yx)/dist(yx)
mean_simple = []
error_simple = []
std_simple = []
z = func(x)


for i in range(count):
        means = np.mean(y[:i+1])
        mean_value.append(means)
        value = np.std(y[:i+1])
        error = value/(i+1)**0.5
        error_value.append(error)
        std.append(value)

        means_simple = np.mean(z[:i+1])*(b-a)
        mean_simple.append(means_simple)
        values_simple = np.std(z[:i+1])
        errors_simple = values_simple/(i+1)**0.5
        error_simple.append(errors_simple)
        std_simple.append(values_simple)

print(f"Importance sampling value is: {mean_value[-1]} and simple sampling value is: {mean_simple[-1]}")
print(f"Importance sampling error value is: {error_value[-1]} and simple sampling error value is: {error_simple[-1]}")
```

```python
py.subplot(1,2,1)
py.xlabel("Steps")
py.ylabel("Integration Value")
py.plot(x_data,mean_value, marker="")
py.title("Importance Sampling")
py.ylim(0.6,0.9)

py.subplot(1,2,2)
py.plot(x_data,mean_simple)
py.xlabel("Steps")
py.ylabel("Integration Value")
py.title("Simple Sampling")
py.ylim(0.6,0.9)

py.suptitle("Different Sampling Types")
py.subplots_adjust(wspace=0.8)
py.show()
#################################################################3
py.subplot(1,2,1)
py.xlabel("Steps")
py.ylabel("Error Value")
py.plot(x_data,error_value)
py.title("Importance Sampling Error")

py.subplot(1,2,2)
py.plot(x_data,error_simple)
py.xlabel("Steps")
py.ylabel("Error Value")
py.title("Simple Sampling Error")

py.suptitle("Errors")
py.subplots_adjust(wspace=0.8)
py.show()
###############################################################
sorted = np.sort(x)
py.scatter(sorted, func(sorted))
py.scatter(sorted, dist(sorted))
py.scatter(sorted, func(sorted)/dist(sorted))
py.legend(["f(x)","p(x)","f(x)/p(x)"])
py.xlabel("x-axis")
py.ylabel("y-axis")
py.show()
```
