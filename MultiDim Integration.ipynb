```python
#Multi Dimenional Integration using MonteCarlos Method
import numpy as np
import matplotlib.pyplot as py
from scipy.stats import norm
```

```python
#There is vectors so expand function first and total variable = total sum of vector variables

def func(x1,x2,x3,y1,y2,y3):
    return np.exp(-(1.5)*(x1**2 + x2**2 )+(x1*x2))#np.exp(-(1.5)*(x1**2 + x2**2 + x3**2 + y1**2 + y2**2+ y3**2)+(x1*y1+ x2*y2 + x3*y3))


n = 100000*2
#x is [o,1]
#limits is (a,b) and a-b = 1
a = -5
b = 5
demo = range(n)
x1 = -5.0 + 10* np.random.uniform(0,1,n)
x2 = -5.0 + 10* np.random.uniform(0,1,n)
x3 = -5.0 + 10* np.random.uniform(0,1,n)
y1 = -5.0 + 10* np.random.uniform(0,1,n)
y2 = -5.0 + 10* np.random.uniform(0,1,n)
y3 = -5.0 + 10* np.random.uniform(0,1,n)

y = func(x1, x2, x3, y1, y2, y3)
error_values = []
mean_values = []
for i in range(n):
    mens = np.mean(y[:i+1])*(b-a)**2
    mean_values.append(mens)
    devi = np.std(y[:i+1]*(b-a)**2)
    error = devi/(1+i)**0.5
    error_values.append(error)

```

```python
py.figure(figsize=(10,5))  
py.grid(True)
#py.axes(projection ="3d")
py.title("Value of integration")
py.scatter(demo,mean_values, marker= ".")
py.show()

py.figure(figsize=(10,5))  
py.grid(True)
py.title("Error")
py.scatter(demo,error_values, marker= ".")
#py.subplots_adjust(wspace=0.8)
py.show()
```
