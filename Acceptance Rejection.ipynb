```python
import numpy as np
import matplotlib.pyplot as py
```

```python
def f(x):
    return (x**3)*np.sin(x)
```

```python
count = 100000
x = np.random.uniform(0,3.5,count)
y = np.random.uniform(0,10,count)
accepted= []
y1 = f(x)
for i in range(count):
    if y1[i]<y[i]:
        accepted.append((x[i],y[i]))
```

```python
x_accepted = [x[0] for x in accepted]
y_accepted = [x[1] for x in accepted]
sorted = np.sort(x_accepted)

x_filtered = sorted#[sorted<= np.pi]
f_filtered = f(x_filtered)

py.scatter(x,y, marker=".")
py.scatter(x_accepted, y_accepted, marker=".")
py.scatter(x_filtered,f_filtered , marker=".")
py.legend(["Accepted Points","Rejected Points","f(x): (x^3)*sin(x)"], loc = 1)
py.xlabel("x-values")
py.ylabel("f(x) or y-values")
py.ylim(0,12)
py.show()
```
