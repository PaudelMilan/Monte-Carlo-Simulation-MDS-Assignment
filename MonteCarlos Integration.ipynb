```python
import numpy as np
import matplotlib.pyplot as py
```

```python
#Integration Using Montecarlos
#Integrate exp(x) over lower limits of 10 and upper limits of 15
a = 10
b = 15
count = 100000
x_data = range(count)
mean_value = []
error_value = []
x = np.random.uniform(a,b,count)
y = np.exp(x)
for i in range(count):
    means = np.mean(y[:i+1])*(b-a)
    mean_value.append(means)
    error = np.std(y[:i+1])/(i+1)**0.5
    error_value.append(error)

py.subplot(1,2,1)
py.title("Integration")
py.xlabel("Random Number Count")
py.ylabel("Integration Value")
py.scatter(x_data,mean_value, marker= ".")
py.subplot(1,2,2)
py.title("Error")
py.xlabel("Random Number Count")
py.ylabel("Error Values")
py.scatter(x_data,error_value, marker= ".")
py.subplots_adjust(wspace=0.8)
py.suptitle(f"Integration of exp(x) with lower and upper limits of {a}, {b} respectively")
py.show()
   
```

```python
squares = y**2
mean_of_numbers = np.mean(y)**2
var = np.mean(squares - mean_of_numbers)
std = var**0.5
```

```python
print("Standard Deviation is:",std)
```
